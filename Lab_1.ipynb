```python
'''
Class and Object

Create a Student class with  :
 1. RollNumber
 2. Name
 3. Marks
Create multiple objects and display their details.
'''
```

```python
class Student:
  def __init__(self,RollNumber,Name,Marks):
    self.RollNumber = RollNumber
    self.Name = Name
    self.Marks = Marks
  def display(self):
    print(self.RollNumber,self.Name,self.Marks)
a=Student(12,'Anil',100)
b=Student(45,'Sumanth',100)
c=Student(89,'Kumar',100)
a.display()
b.display()
c.display()
```
