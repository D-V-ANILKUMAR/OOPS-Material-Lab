```python
'''
Lab 3 - Parameterized Constructor
Create a Book class containing:
Title
Author
Price
Initialize values through a parameterized constructor.
'''
```

```python
class Book:
  def __init__(self, title, author, price):
    self.title = title
    self.author = author
    self.price = price
  def display(self):
    print(self.title, self.author, self.price)
ob=Book('THE BOOK','ANIL',100)
ob.display()
```
