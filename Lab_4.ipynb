```python
'''
Lab 4–Instance Methods
Create a Rectangle class.
Implement methods to calculate:
Area
Perimeter
'''
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2*(self.length+self.width)
ob=Rectangle(2,4)
print("Area :",ob.area())
print("Perimeter :",ob.perimeter())
```
