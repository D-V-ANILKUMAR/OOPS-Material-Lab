```python
'''
Lab 2 - Constructors
Create an Employee class and initialize employee information using a constructor.
Display employee details.
'''
```

```python
class Employee:
  def __init__(self,ID,Name,Salary):
    self.ID = ID
    self.Name = Name
    self.Salary = Salary
  def display(self):
    print(self.ID,self.Name,self.Salary)
a=Employee(12,'Anil',100)
a.display()
```
