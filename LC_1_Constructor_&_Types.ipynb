```python
# Constructor
# A Constructor is a default Method present inside a Class.
#	In Python constructor is defined using __init__() Method.
#	It is automatically called when object is Created.
#	In Python Constructors are two types. They are :
#    1.	Default Constructor or Non-Parameterized Constructor
#    2.	Parameterized Constructor.
```

```python
class student:
  def display(self):
    print(self.name)
  def set(self):
    print("Anil")
  def __init__(self,name='Anil'):
    self.name=name
    print(self.name)
ob=student("Pranadeep")
ob.display()
```

```python
#	Reference Pointer (self) :
#	Self is the Default Keyword Present in the Constructor.
# It holds the address of the Current Object.
```

```python
# 1. Default Constructor :
#	If a Constructor is Implemented without using any Parameters other than Self Keyword is known as Default Constructor.
```

```python
# Default Constructor
class student:
  def __init__(self):
    print('Hello! I am Default Constructor')
ob=student()
```

```python
#	2. Parameterized Constructor :
#	If a Constructor is Implemented by using One or More Parameters Including Self Keyword is known as Parameterized Constructor.
```

```python
# Parameterized Constructor
class student:
  def __init__(self,name):
    self.name=name
    print('Hello! ',self.name)
    print('I am Parameterized Constructor')
ob=student('Anil')
```
