```python
class car:
  def __init__(self,brand,model,year):
    self.brand=brand
    self.model=model
    self.year=year
  def get(self):
    print(self.brand,self.model,self.year)
ob=car('Maruti','DZire',2000)
ob.brand='Innova'
ob.model="Crysta"
ob.year=2026
ob.get()
```
